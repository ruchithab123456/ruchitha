```python
import pandas as pd
import matplotlib.pyplot as m
d={'mpg':[18,15,18,16,17],
   'cylinders':[8,8,6,4,8],
   'displacement':[307,350,318,304,302],
   'horsepower':[130,165,150,150,140],
   'weight':[3504,3693,3436,3433,3449],
   'acceleration':[12.0,11.5,11.0,12.0,10.5],
   'model_year':[70,71,70,80,70],
   'origin':[1,1,1,1,1],
   'carname':['chervolt chevele maibu','buick skylack 320','plymouth stalite','amc rebel set','ford torino']}
df=pd.DataFrame(d)
sa=df.describe()
cy=df[df['cylinders']==8]
yr=df.groupby('model_year')['model_year'].count()
print("Statistical:\n",sa)
print("\nCylinders:\n",cy)
print("\nYears:\n",yr)
```

```python
import pandas as pd
import matplotlib.pyplot as m
import seaborn as s

data=pd.read_csv("E:\\AIruchi\\mtcars.csv")

#HISTOGRAM
mpg=data['mpg']
m.hist(mpg,bins='auto',color='k',edgecolor='c')
m.xlabel('MPG')
m.ylabel('frequency')
m.title('Frequency distribution of mpg')
m.show()

#SCATTERPLOT
wt=data['wt']
a=range(len(data))
m.scatter(a,mpg,color='k',label='mpg')
m.scatter(a,wt,color='c',label='wt')
m.title('Relation between weight and car')
m.legend()
m.show()

#BARPLOT
c=data['am'].value_counts()
co=['k','c']
m.bar(c.index,c.values,color=co,width=0.3)
m.xlabel('Transmission type')
m.ylabel('No.of values')
m.title('Distribution of transmission type of car')
m.show()

#BOXPLOT
s.boxplot(mpg,color='c')
m.xlabel('MPG')
m.ylabel('No.of values')
m.show()
```
